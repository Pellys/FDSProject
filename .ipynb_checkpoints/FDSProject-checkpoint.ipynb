```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # sets up plotting under plt
import seaborn as sns # sets up styles and gives us more plotting options


#also divided per platform
```

```python
ds= pd.read_csv("games_sales_clean.csv", header="infer")
ds
```

```python
#Global and regional number of games produced per year
sales_platform = ds.groupby(['Platform'])['Year_of_Release'].agg(Year_of_release='sum')

plt.figure(figsize=(16,8))

sns.barplot(x=sales_platform.index, y=sales_platform['Year_of_Release'])
```
