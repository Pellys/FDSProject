# regression

```python
games= pd.read_csv("games_sales_clean.csv", header="infer")
games=games[games.Year_of_Release >=2000]
```

```python
y=np.array(games.groupby(["Year_of_Release"]).Global_Sales.sum())                #creating ad array with global sales(y=target of prediction)
```

```python
x=np.array(np.arange(2000,2017))                #creating an array with our features
```

##  linear regression

We can see, just looking at the data, that a linear regression will not well describe our data. Here's an example tring to predict the future Global sales using the least squares method

### least squares method

$y= T(x)+S(x)+A(x)$

where T(x) it's the trend component, S(x) it's the seasonal component and A(x) it's the Alatory component. Since there is no seasonal component(just look at the data) and the alatory component it's just a little adjustment, we can just look at the trend component and see if a linear one can fit the data

$T(x)=a + bx$

$b= \frac{cov(x,y)}{var(x)}$

$a= y_{mean} - bx_{mean}$

```python
y_mean= np.mean(y)
y_mean
```

```python
x_mean=np.mean(x)
x_mean
```

```python
cov_xy=np.cov(x,y)[0][1]
var_x=np.cov(x,y)[0][0]
```

```python
'''
covxy=0
for i in range(x.shape[0]):
    covxy+= ((x[i]-x_mean)*(y[i]-y_mean))
covxy/x.shape[0]
'''
```

```python
b=cov_xy/var_x
a=y_mean - b*x_mean

T_x=a + b*x
```

```python
plt.figure(figsize=(15, 6))
plt.plot(x,y)
plt.plot(x,T_x)
plt.legend(["Global Sales","Trend"],fontsize=15)
plt.ylabel('Sales(Mln)',fontsize=30)
plt.xlabel("years",fontsize=30)
plt.show()
```

as we can see there is no point on using a linear regression. let's try to use a polynomial one

# polynomial regression

```python
#my h(theta)
def lin_f(theta, x):
    
    return theta[1]*x + theta[0]

def pol2_f(theta, x):
    
    return theta[2]*x**2 + theta[1]*x + theta[0]

def pol3_f(theta, x):
    
    return theta[3]*x**3 + theta[2]*x**2 + theta[1]*x + theta[0]

```

```python
#gradient of my cost function
def grad_costf(theta,x,y,h):
    G=np.zeros(len(theta))
    for i in range (len(x)):
        for j in range(len(theta)):
            G[j] += (h(theta,x[i])-y[i])/(len(x)) * x[i]**j            #if we have 2° or higher order terms we need **j
    return G        
```

```python
# gradient descent
def GD(theta,x,y,h,alpha=0.001,eps=0.000001):
    grad_cost_history=[]
    while np.linalg.norm(alpha*grad_costf(theta, x, y, h))>=eps:
        theta = theta - alpha*grad_costf(theta, x, y, h)
        
    return theta    
    
```

## Global

```python
y=np.array(games.groupby(["Year_of_Release"]).Global_Sales.sum())                #creating ad array with global sales(y=target of prediction)
x=np.array(np.arange(0,17))                #creating an array with our features
```

#### 1st order regression

```python
#initialize theta0
theta0=np.zeros(2)
#use GD
theta_final_1=GD(theta0,x,y,lin_f)
print(theta_final_1)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(x+2000,y)
plt.plot(x+2000,lin_f(theta_final_1,x))
plt.legend(["Global Sales","Trend"],fontsize=15)
plt.ylabel('Sales(Mln)',fontsize=30)
plt.xlabel("years",fontsize=30)
plt.show()
```

#### 2nd order regression

```python
#initialize theta0
theta0=np.zeros(3)
#use GD
theta_final_2=GD(theta0,x,y,pol2_f,alpha=0.0001)
print(theta_final_2)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(x+2000,y)
plt.plot(x+2000,pol2_f(theta_final_2,x))
plt.legend(["Global Sales","Trend"],fontsize=15)
plt.ylabel('Sales(Mln)',fontsize=30)
plt.xlabel("years",fontsize=30)
plt.show()
```

#### 3rd order regression

```python
#initialize theta0
theta0=np.zeros(4)
#use GD
theta_final_3=GD(theta0,x,y,pol3_f,alpha=0.0000001)
print(theta_final_3)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(x+2000,y)
plt.plot(x+2000,pol3_f(theta_final_3,x))
plt.legend(["Global Sales","Trend"],fontsize=15)
plt.ylabel('Sales(Mln)',fontsize=30)
plt.xlabel("years",fontsize=30)
plt.show()
```
